# Fungsi Tranformasi
Buat fungsi untuk transformasi dengan isi meliputi
    # import data
    # hilangkan missing value
    # gabungkan data
    # transformasi kolom yang sesuai

```python
#buat fungsi
def transform_data (bps_file, gee_file) :
    bps = pd.read_csv(bps_file)
    gee = pd.read_csv(gee_file)
    bps.dropna(axis="columns", how='all', inplace=True)
    bps = bps.dropna(axis="index", thresh=2, inplace=False)
    bps.fillna({'Nila (2023)': 'NO TANGKAP'}, inplace=True)
    bps_new = bps.merge(gee, how='left', on='Provinsi', suffixes=('_bps','_gee'))
    bps_new = bps_new.groupby(['Nila (2023)']).agg({'Provinsi':list, 'Suhu rata-rata':'mean'}).reset_index()
    bps_new.rename(columns={"Nila (2023)":"Ikan Nila 2023"},inplace=True)

    return bps_new
```

```python
# jalankan fungsi
bps_new = transform_data(bps,gee)
bps_new.head()
```

```python
# Save to CSV
bps_new.to_csv("new.csv", index=False, sep=';', decimal=',', encoding='utf-8')
```

# Pipeline
Buat fungsi Fungsi Pipeline untuk menjalankan Fungsi Pembersihan dan Transformasi
1. pipeline baca data
2. pipeline hapus baris null
3. pipeline misssing values
4. pipeline merge data

```python
def pipeline_baca_data():
    bps = 'produksi_perikanan_budidaya_final.csv'
    gee = 'suhu_perbandingan.csv'
    bps_df = pd.read_csv(bps)
    gee_df = pd.read_csv(gee)
    return bps_df, gee_df
```

```python
def pipeline_hapus_null(bps_df, gee_df):
    bps_df = bps_df.dropna(how='all')
    gee_df = gee_df.dropna(how='all')
    return bps_df, gee_df
```

```python
def pipeline_missing_values(bps_df):
    bps_df.fillna({'Nila (2023)': 'NO TANGKAP'}, inplace=True)
    return bps_df
```

```python
def pipeline_merge_data(bps_df, gee_df):
    merged_df = pd.merge(bps_df, gee_df, on='Tahun', how='inner')
    return merged_df
```

```python
def pipeline():
    bps_df, gee_df = pipeline_baca_data()
    bps_df, gee_df = pipeline_hapus_null(bps_df, gee_df)
    bps_df = pipeline_missing_values(bps_df)
    final_df = pipeline_merge_data(bps_df, gee_df)
    return final_df
```
