```python
import pandas as pd

# Membaca file dengan 2 baris header
file_path = "/content/drive/MyDrive/produksi_ikan_bersih.csv"
df_raw = pd.read_csv(file_path, header=[0,1])

# Gabungkan multi-index kolom menjadi satu nama kolom
df_raw.columns = ['_'.join([str(a).strip().lower().replace(" ", "_")
                            for a in col if str(a).strip() != '' and str(a).lower() != 'nan'])
                  for col in df_raw.columns]

# Hapus kolom kosong/unnamed
df_raw = df_raw.loc[:, ~df_raw.columns.str.contains('^unnamed')]

# Hapus baris kosong
df_raw = df_raw.dropna(how='all')

# Ubah semua nilai '-' jadi NaN agar bisa diolah
df_raw = df_raw.replace('-', pd.NA)

# Tampilkan info
print("=== Info Data ===")
print(df_raw.info())

# Profil awal
print("\n=== Deskripsi Data Numerik ===")
print(df_raw.describe())

print("\n=== Deskripsi Data Kategorikal ===")
print(df_raw.describe(include='object'))

# Cek missing values
print("\n=== Cek Nilai Kosong ===")
print(df_raw.isnull().sum())

# Cek duplikasi
print("\n=== Cek Duplikasi ===")
print(df_raw.duplicated().sum())

# Preview data
print("\n=== Preview Data Setelah Dibersihkan ===")
print(df_raw.head())
```
